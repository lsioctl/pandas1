```python

df = pd.read_csv(
    'bikes.csv',
    # Note: the encoding is not anymore latin1, we have real ? in ASCII
    encoding='latin1',
    sep=';',
    parse_dates=['Date'],
    # the date in the csv has day first and not the month first
    dayfirst=True,
    index_col=['Date']
)
```

```python
df
```
